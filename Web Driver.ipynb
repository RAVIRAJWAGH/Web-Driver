```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time

def refresh_chrome():
    # Replace 'path/to/chromedriver' with the path to your chromedriver executable
    driver_path = r"C:\Users\Parimal\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"

    # Specify the existing Chrome profile directory
    user_data_dir = r"C:\Users\Parimal\AppData\Local\Google\Chrome\User Data\Profile 2"

    # Create Chrome options
    chrome_options = Options()
    chrome_options.add_argument(f"--user-data-dir={user_data_dir}")

    # Create a Chrome browser instance
    service = ChromeService(driver_path)
    driver = webdriver.Chrome(service=service, options=chrome_options)

    try:
        while True:
            try:
                # Open a website of your choice
                driver.get("https://expert.chegg.com/qna/authoring/answer")

                # Wait for 2 minutes before refreshing
                time.sleep(5)

                # Find the element with XPath
                noq = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, '//*[@id="__next"]/main/div/div/div[1]/h4'))
                )
                print(noq.text)
                print(type(noq.text))
                # Check if "Dear Expert!" is present in the text of the element
                if "Dear Expert!" in noq.text:
                    print("The text 'Dear Expert!' is present.")
                else:
                    print("The text 'Dear Expert!' is not present.")

            except KeyboardInterrupt:
                # Handle keyboard interrupt (Ctrl+C) to gracefully exit the script
                print("Script terminated by user.")
                break
    finally:
        # Close the browser window
        driver.quit()

if __name__ == "__main__":
    refresh_chrome()
```
